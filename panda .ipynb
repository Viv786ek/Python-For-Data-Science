```python
# to find the missing value

loan.isnull().sum()
```

```python
loan.drop('Credit_History',axis=1,inplace=True)
```

```python
loan['LoanAmount'].fillna(loan['LoanAmount'].mean(),inplace=True)
```

```python
loan.isnull().sum()
```

```python
loan['Loan_Amount_Term'].fillna(loan['Loan_Amount_Term'].mode(),inplace=True)
```

```python
loan['Loan_Amount_Term'].fillna(loan['Loan_Amount_Term'].mode()[0],inplace=True)
```

```python
loan['Self_Employed'].value_counts()
```

```python
loan['Self_Employed'].fillna(('No'),inplace=True)
```

```python
loan['Dependents'].str.rstrip('+')  # rstrip is string function   using ->>> str
```

```python
loan['Dependents']=loan['Dependents'].str.rstrip('+')
```

```python
loan.isnull().sum()
```

```python
loan['Dependents'].fillna(method="ffill",inplace=True)  # better than the previoue one  -->> method="ffill"---> forword fill
```

```python
loan.isnull().sum()
```

```python
loan['Married'].fillna(loan['Married'].mode()[0],inplace=True)
```

```python
loan.isnull().sum()
```

```python
loan['Gender'].fillna(loan['Gender'].mode()[0],inplace=True)
```

```python
loan.isnull().sum()
```

```python
# convert from string to  numeric using label encoder
from sklearn.preprocessing import LabelEncoder
```

```python
lec=LabelEncoder()
```

```python
lec.fit(loan['Married'])
```

```python
loan['Married']=lec.fit_transform(loan['Married'])
```

```python
loan['Education']=lec.fit_transform(loan['Education'])
```

```python
loan.head()
```

```python
# convert string to numeric using get_dummies   ---> it is a function of panda
Gender2=pd.get_dummies(loan['Gender'])
```

```python
Gender2
```

```python
pd.concat([loan,Gender2],axis=1)
```

```python
loan=pd.concat([loan,Gender2],axis=1)
```
